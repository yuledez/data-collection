## Deliverable 1: Scrape Titles and Preview Text from Mars News

```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup
import json
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

1. Use automated browsing to visit the [Mars news site](https://static.bc-edx.com/data/web/mars_news/index.html). Inspect the page to identify which elements to scrape.

      > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools.

```python
# Visit the Mars news site
url = 'https://static.bc-edx.com/data/web/mars_news/index.html'
browser.visit(url)
```

### Step 2: Scrape the Website

Create a Beautiful Soup object and use it to extract text elements from the website.

```python
# Create a Beautiful Soup object
# Collect the HTML from the browser
html = browser.html

# Print the HTML content (for debugging)
print(html)

# Parse the HTML with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Save the main area of the web page to a variable
main_area = soup.find("section", class_="image_and_description_container")

# Check if main_area is found
if main_area is None:
    print("Main area not found")
else:
    # Extract all the article containers
    all_articles = main_area.find_all("div", class_="col-md-8")

    # Check if any articles were found
    if not all_articles:
        print("No articles found")
    else:
        # Create an empty list to store the dictionaries
        summary_list1 = []

        # Loop through the article containers
        # Extract the title and preview text from the elements
        # Store each title and preview pair in a dictionary
        # Add the dictionary to the list
        for article1 in all_articles:
            # Debugging: Print the article element
            #print(article1.prettify())

            # Collect the article title
            title1 = article1.find("div", class_="content_title")
            if title1:
                title1 = title1.text.strip()
            else:
                title1 = "No Title Found"

            # Collect the article preview
            preview1 = article1.find("div", class_="article_teaser_body")
            if preview1:
                preview1 = preview1.text.strip()
            else:
                preview1 = "No Preview Found"       
                    
            # Create the summary dictionary
            summary_dict1 = {
                "Title": title1,
                "Preview": preview1
            }
            # Append the summary dictionary to the list
            summary_list1.append(summary_dict1)
        
        # Print the list to confirm success
        print(summary_list1)

        # Export the summary list to a JSON file
        with open('mars_news.json', 'w') as json_file:
            json.dump(summary_list1, json_file, indent=4)

browser.quit()
```
